```python
%matplotlib inline
import cv2
import os
from matplotlib import pyplot as plt
import random
```

```python
path_to_data = "/mnt/Data/Documents/Praca_magisterska/dataset/dataset-master/unlabeled"
files_in_data = os.listdir(path_to_data)
paths_in_data = [os.path.join(path_to_data, x) for x in files_in_data]
dirs = [x for x in paths_in_data if os.path.isdir(x)]
```

```python
dirs[0].split("/")
```

```python
for d in dirs:
    print(d)
    labels_file = os.path.join(path_to_data, d + ".csv")
    print(labels_file)
    labels = {}
    with open(labels_file, "r") as lf:
        for l in lf:
            split = l.split(",")
            labels[split[0]] = (split[1], split[2], split[3], split[4], split[5], split[6], split[7], split[8])
    
    files = [os.path.join(d, x) for x in os.listdir(d)]
    rand_samples = [files[i] for i in sorted(random.sample(range(len(files)), 10))]
    for s in rand_samples:
        print(s)
        img = cv2.imread(s)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        try:
            l = labels[s.split("/")[-1]]
            cv2.rectangle(img, (int(l[2]), int(l[3])), (int(l[6]), int(l[7])), (0,255,0),2)
            plt.figure()
            plt.imshow(img)
        except:
            print("skipped")

        
        
```
